## metadata analysis by season

```python
def get_bag_of_words(df, column_name, low_limit=5, k=20):
    bag_of_words_dict= dict()

    for subject in df.loc[df[column_name]>low_limit]['Subjects']:
        try:
            for word in subject.split(","):
                if word in bag_of_words_dict.keys():
                    bag_of_words_dict[word] = bag_of_words_dict[word]+1
                else:
                    bag_of_words_dict[word] = 1
        except:
            continue
                
    print(sorted(bag_of_words_dict.items(), key=operator.itemgetter(1), reverse=True)[:k])
```

```python
get_bag_of_words(df_overindex, 'summer')
```

```python
get_bag_of_words(df_overindex, 'fall')
```

```python
get_bag_of_words(df_overindex, 'winter')
```

```python
get_bag_of_words(df_overindex, 'spring')
```
