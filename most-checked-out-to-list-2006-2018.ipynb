```python
import pandas as pd
import matplotlib.pylab as plt
from datetime import datetime
plt.style.use('fivethirtyeight')
%matplotlib inline
```

```python
df = pd.read_csv('./seattle-library-checkout-records/most_checked_out_all.csv')
print(len(df))
```

```python
df.head()
```

```python
def top_list(year, format_subgroup, k=5):
    top = df[df['year']==year][df['format_subgroup'].apply(lambda x: x in format_subgroup)].sort_values(by = 'num_checkouts', ascending = False)
    return top[['author','title','publication_year','num_checkouts']].head(k)
```

```python
top_list(2018, ['Video Disc','Audio Disc', 'Music Score'], k=10)
```

```python
top_list(2018, ['Book'], k=20).drop_duplicates()
```

```python
for year in range(2006,2019):
    print('=======================================================================================')
    print(year)
    for format in [['Book'], ['Video Disc','Audio Disc', 'Music Score']]:
        print(format)
        print('=========')
        print(top_list(year, format, k=1))
    print('=======================================================================================')
```
