```python
import pandas as pd
import matplotlib.pylab as plt
plt.style.use('fivethirtyeight')
%matplotlib inline
```

```python
most_checked = pd.read_csv('./seattle-library-checkout-records/most_checked_out_all.csv')
print(len(most_checked))
```

```python
most_checked.head()
```

```python
plt.title('Number of Total  Checkouts', fontsize = 16)
plt.bar(range(13), most_checked.groupby('year')['num_checkouts'].sum()/100000)
plt.bar(range(13), most_checked[most_checked['format_subgroup']=='Book'].groupby('year')['num_checkouts'].sum()/100000, color= 'coral', label='Books')

plt.xticks([x-0.3 for x in range(13)], [str(x) for x in range(2006, 2019)], rotation='45')
plt.ylabel('Number of Checkouts, in 100K', fontsize = 16)
plt.tick_params(labelsize = 16)
plt.legend(loc='best')
```

```python
plt.title('% of Checkouts that are Books', fontsize = 16)
plt.bar(range(len(most_checked['year'].unique())),most_checked[most_checked['format_subgroup']=='Book'].groupby('year')['num_checkouts'].sum() / most_checked.groupby('year')['num_checkouts'].sum()*100)

plt.xticks([x-0.3 for x in range(13)], [str(x) for x in range(2006, 2019)], rotation='45')
plt.ylabel('%', fontsize = 16)
plt.tick_params(labelsize = 16)
#plt.legend(loc='best')
```

```python
plt.title('Variety of Unique Items', fontsize = 16)
plt.bar(range(13), most_checked.groupby('year')['num_checkouts'].size()/1000)
plt.bar(range(13), most_checked[most_checked['format_subgroup']=='Book'].groupby('year')['num_checkouts'].size()/1000, color= 'coral', label='Books')

plt.xticks([x-0.3 for x in range(13)], [str(x) for x in range(2006, 2019)], rotation='45')
plt.ylabel('Number of Checkouts, in 1K', fontsize = 16)
plt.tick_params(labelsize = 16)
plt.legend(loc='best')
```

```python
plt.title('% of Unique Items that are Books', fontsize = 16)
plt.bar(range(13), most_checked[most_checked['format_subgroup']=='Book'].groupby('year')['num_checkouts'].size() / most_checked.groupby('year')['num_checkouts'].size()*100)

plt.xticks([x-0.3 for x in range(13)], [str(x) for x in range(2006, 2019)], rotation='45')
plt.ylabel('%', fontsize = 16)
plt.tick_params(labelsize = 16)
#plt.legend(loc='best')
```
